```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
# 필요한 데이터셋 불러오기
iris = load_iris()
```

```python
# 데이터프레임 만들기
df = pd.DataFrame(iris['data'],columns = iris['feature_names'])
df['target'] = iris['target']
df.head()
```

```python
df.iloc[:,:4]
```

```python
# 데이터 셋 분할
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,:4],df['target'], stratify = df['target'])
print('train',x_train.shape, y_train.shape)
print('valid',x_test.shape, y_test.shape)
```

```python
# 모델 정의
knn = KNeighborsClassifier()
# 학습
knn.fit(x_train,y_train)
```

```python
# 검증
prediction = knn.predict(x_test)
(prediction == y_test).mean()
```

```python
knn.score(x_test,y_test)
```

```python
for k in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(x_train,y_train)
    score = knn.score(x_test,y_test)
    print('k: %d, accuracy: %.2f'%(k,score*100))
```

# Moon Data

```python
from sklearn.datasets import make_moons # 달 모양 데이터셋
# 매번 같은 데이터가 나오도록 seed 고정
random_seed = 2020

# 데이터 개수
n_samples = 1000

feature,target = make_moons(n_samples=n_samples, noise=0.2,random_state=random_seed)
```

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['royalblue','orangered'])

fig = plt.figure(figsize=(12,12))
plt.scatter(feature[:,0],feature[:,1], c=target, edgecolors = 'k', cmap=cmap)
plt.show()
```

# Decision Tree

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(feature,target,test_size=0.2,random_state=1)

from sklearn import tree
from sklearn.metrics import accuracy_score

maxdepths = [2,3,4,5,6,7,8,9,10,15,20,30,40,50]

trainAcc = np.zeros(len(maxdepths))
testAcc = np.zeros(len(maxdepths))

index = 0
for depth in maxdepths:
    clf = tree.DecisionTreeClassifier(max_depth=depth)
    clf.fit(X_train,Y_train)
    Y_predTrain = clf.predict(X_train)
    Y_predTest = clf.predict(X_test)
    trainAcc[index] = accuracy_score(Y_train,Y_predTrain)
    testAcc[index] = accuracy_score(Y_test,Y_predTest)
    index+=1
plt.plot(maxdepths,trainAcc,'ro-',maxdepths,testAcc,'bv--')
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000','#00FF00','#0000FF'])

# k가 클수록 local=>global
for k in [1,50]:
    clf = KNeighborsClassifier(n_neighbors = k)
    clf.fit(X_train,Y_train)
    
    x_min,x_max = X_train[:,0].min() -1, X_train[:,0].max()+1
    y_min,y_max = X_train[:,1].min() -1, X_train[:,1].max()+1
    xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),
                       np.arange(y_min,y_max,.02))
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(10,10))
    plt.pcolormesh(xx,yy,Z,cmap=cmap_light)
    
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=cmap_bold)
plt.show()
```

```python
from tqdm import tqdm
from sklearn.model_selection import cross_val_score

k_list = [i for i in range(1,100) if i%2 == 1]
cv_scores = []

for k in tqdm(k_list):
    # p1=> manhattan_distance, p2=> euclid_distance, p3=> minkowsi_distance
    classifier = KNeighborsClassifier(n_neighbors = k, p=2)
    scores = cross_val_score(classifier,X_train,Y_train,cv=10,scoring='accuracy')
    cv_scores.append(scores.mean())
for i in range(len(cv_scores)):
    print(i*2+1,cv_scores[i])
```

```python
# manhattan_distance
manhattan=[]
for k in tqdm(k_list):
    classifier = KNeighborsClassifier(n_neighbors = k, p=1)
    scores = cross_val_score(classifier,X_train,Y_train,cv=10,scoring='accuracy')
    manhattan.append(scores.mean())
```

```python
# minkowsi_distance
minkowsi=[]
for k in tqdm(k_list):
    classifier = KNeighborsClassifier(n_neighbors = k, p=3)
    scores = cross_val_score(classifier,X_train,Y_train,cv=10,scoring='accuracy')
    minkowsi.append(scores.mean())
```
